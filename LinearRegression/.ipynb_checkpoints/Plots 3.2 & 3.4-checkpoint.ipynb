```python
# Sample data (Advertising data)
x = ad_data.TV
y = ad_data.sales

# Generate a range of beta_0 and beta_1 values
beta_0_vals = np.linspace(5, 9, 200)  # Range for beta_0
beta_1_vals = np.linspace(0.03, 0.06, 200)  # Range for beta_1

# Create a grid for beta_0 and beta_1
B0, B1 = np.meshgrid(beta_0_vals, beta_1_vals)

# Calculate RSS for each pair of (beta_0, beta_1)
RSS = np.zeros_like(B0)
for i in range(len(beta_0_vals)):
    for j in range(len(beta_1_vals)):
        beta_0 = B0[i, j]
        beta_1 = B1[i, j]
        RSS[i, j] = np.sum((y - (beta_0 + beta_1 * x))**2)

# Plot the contour
plt.contour(B0, B1, RSS, colors='blue')
plt.xlabel(r'$\beta_0$')
plt.ylabel(r'$\beta_1$')
plt.title('RSS Contours')
plt.scatter([7], [0.05], color='red')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Sample data (Advertising data)
x = np.array([1, 2, 3, 4, 5])  # Predictor
y = np.array([2, 4, 5, 4, 5])  # Response

# Generate a range of beta_0 and beta_1 values
beta_0_vals = np.linspace(5, 9, 100)  # Range for beta_0
beta_1_vals = np.linspace(0.03, 0.06, 100)  # Range for beta_1

# Create a grid for beta_0 and beta_1
B0, B1 = np.meshgrid(beta_0_vals, beta_1_vals)

# Calculate RSS for each pair of (beta_0, beta_1)
RSS = np.zeros_like(B0)
for i in range(len(beta_0_vals)):
    for j in range(len(beta_1_vals)):
        beta_0 = B0[i, j]
        beta_1 = B1[i, j]
        RSS[i, j] = np.sum((y - (beta_0 + beta_1 * x))**2)

# Create a 3D plot
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
surface = ax.plot_surface(B0, B1, RSS, cmap='Blues', edgecolor='k', alpha=0.8)

# Add a point for the minimum RSS (adjust based on actual minimum values)
min_beta_0 = 7
min_beta_1 = 0.05
min_RSS = np.sum((y - (min_beta_0 + min_beta_1 * x))**2)
ax.scatter(min_beta_0, min_beta_1, min_RSS, color='red', s=100, label='Minimum RSS')

# Labels and title
ax.set_xlabel(r'$\beta_0$', labelpad=10)
ax.set_ylabel(r'$\beta_1$', labelpad=10)
ax.set_zlabel('RSS', labelpad=10)
ax.set_title('3D Plot of RSS vs $\\beta_0$ and $\\beta_1$')

# Show legend and plot
ax.legend()
plt.show()
```
